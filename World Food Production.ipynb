```python
from sklearn.cluster import KMeans
sse = []
for k in range(1,40):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df1_scaled)
    sse.append(kmeans.inertia_)
```

```python
sn.set()
plt.figure(figsize=(6, 6))
plt.plot(range(1, 40), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of Squared Distances (SSE)')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
colors = ['red', 'blue', 'green', 'cyan', 'magenta', 'yellow', 'black', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'lime', 'navy']

# Create a grid of subplots for each pair of features
fig, axs = plt.subplots(13, 13, figsize=(20, 20))

# Loop through each pair of features and each cluster
for i in range(13):
    for j in range(13):
        # Plot each cluster with a different color
        for cluster_id in range(15):
            axs[i, j].scatter(df1_scaled[y == cluster_id, i], df1_scaled[y == cluster_id, j], s=10, c=colors[cluster_id], label=f'cluster{cluster_id+1}')
        axs[i, j].set_xlabel(f'Feature {i+1}')
        axs[i, j].set_ylabel(f'Feature {j+1}')
        axs[i, j].legend()

plt.tight_layout()
plt.show()
```

```python
kmeans = KMeans(n_clusters=15,init="k-means++",random_state=0)
y = kmeans.fit_predict(df1_scaled)
print(y)
```

```python
import matplotlib.pyplot as plt

# List of colors to use for the clusters
colors = ['red', 'blue', 'green', 'cyan', 'magenta', 'yellow', 'black', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'lime', 'navy']

# Plot each cluster with a different color
plt.figure(figsize=(10, 6))

# Choose two features for the scatter plot. Here, using the first two columns as an example.
feature_x_index = 0
feature_y_index = 1

for i in range(15):
    plt.scatter(df1_scaled[y == i, feature_x_index], df1_scaled[y == i, feature_y_index], s=50, c=colors[i], label=f'Cluster {i+1}')

plt.xlabel(f'Feature {feature_x_index + 1}')
plt.ylabel(f'Feature {feature_y_index + 1}')
plt.title('K-Means Clustering')
plt.legend()
plt.show()
```

```python
k_optimal = 15

kmeans = KMeans(n_clusters=k_optimal, random_state=42)
kmeans.fit(df1_scaled)

df1['Cluster'] = kmeans.labels_

print(df1.head())

sn.pairplot(df1, hue='Cluster', vars=features, palette='tab10')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Reduce dimensionality to 2 dimensions using PCA
pca = PCA(n_components=2)
df1_pca = pca.fit_transform(df1_scaled)

# List of colors to use for the clusters
colors = ['red', 'blue', 'green', 'cyan', 'magenta', 'yellow', 'black', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'lime', 'navy']

# Plot each cluster with a different color
plt.figure(figsize=(8, 8))
for cluster_id in range(15):
    plt.scatter(df1_pca[y==cluster_id, 0], df1_pca[y==cluster_id, 1], s=50, c=colors[cluster_id], label=f'cluster{cluster_id+1}')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-Means Clustering with 15 Clusters (PCA Visualization)')
plt.legend()
plt.show()
```
